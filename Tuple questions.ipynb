### tuple slicing

```python
t = (2,3,4,6,3,1)
print(t[0:5])
print(t[5:])    
print(t[1:])
print(t[:4])
print(t[-3:-4])   #empty list
print(t[-4:-3])   #first index element
```

### convert list into tuple

```python
l = [323,32,332,2323,23]
t= tuple(l)
print(t)
```

### remove an element from tuple

```python
t= (22,21,13,31)
list1 = list(t)
list1.remove(22)
t = tuple(list1)
print(t)
```

```python
t = ("apple","banana","grapes")
convert = list(t)
convert.remove("apple")
t = tuple (convert)
print(t)
```

### sort a tuple

```python
t = (2,4,5,7,3,1,4,5,6,0)
print(t.sort)             # sort does not work in tuple
```

```python
t = (2,4,5,7,3,1,4,5,6,0)
sort = (sorted(t))  # sorted function works for list, tuples and string
tuple(t)            # sorted function returns a list,so convert it to tuple
print(sort)
```

### common elements bw two tuples

```python
t1 = (2,2,34,3,2)
t2 = (2,2,3,4,2,1)
common = (set(t1)&set(t2))  #set stores unique values and support intersection operation
print (tuple(common))       #again convert it to tuple
```

### find second largest element

```python
t = ( 2,2,3,4,4,6,3,2,1)
sort = tuple(list(t))         #converting into list
sort = sorted(sort)          #last second element will be the largest
print(sort[-2])
```

```python
t = (4, 1, 7, 3)
unique = tuple(set(t))       #by converting it into set
unique = sorted(unique)
print(unique[-2])
```
