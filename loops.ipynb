#### Printing no. from 1 to 5

```python
for i in range(1, 6): #starts from 1 ends at 5
    print(i)
```

```python
for i in range(3): #will print 3 time the given value
    print("Hello")
                      
```

#### even no. from 2 to 10

```python
for i in range(2, 11, 2):
    print(i)
```

### Loops are of two types while loop and for loop

```python
i = 1         #starts from 1
while i <= 5: #goes until 5
    print(i)
    i += 1    #inc with 1
```

```python
for i in range(1, 11):
    print(i)
```

#### print square from 1 to 5

```python
for i in range(1,6):
    print(i*i)
```

#### print every character of the word

```python
word = "Python"

for ch in word:
    print(ch)

```

#### reverse counting

```python
i = 10
while i >= 1:
    print(i)
    i -=1
```

#### multiplication of a no. n

```python
i = 1
while i <= 10:
     print(3*i)
     i +=1
```

#### print elements from list 

```python
numbers = [10, 20, 30, 40, 50]

for item in numbers:
    print(item)
```

```python
numbers = [10, 20, 30, 40, 50]

for i in range(len(numbers)):
    print(numbers[i])
```

```python
numbers = [10, 20, 30, 40, 50]
i = 0
while i < len(numbers):
     print(numbers[i])
     i +=1
```

#### print only positive no.

```python
data = [-5, 3, -2, 7, 0]

for x in data:
    if x > 0:
        print(x)
```

#### stop loop when no. reaches 11

```python
for i in range(1, 100):
    if i == 12:
        break
    print(i)
```
