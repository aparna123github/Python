## Types of args

### Positional arg - order matters here

```python
def subtract(a, b):
    print(a - b)

subtract(10, 5)
```

```python
def subtract(a, b):
    print(a - b)

subtract(5,10)
```

### Keyword args - order doe not matter here

```python
subtract(b=5, a=10)
```

```python
subtract(b=550, a=100)
```

### Variable length args 

```python
# *args - non keyword args
def add_numbers(*args):
    print(sum(args))

add_numbers(1, 2, 3, 4)
```

```python
# *kwargs  - keyowrd args
def student_info(**kwargs):
    print(kwargs)

student_info(name="Alex", age=20)
```

```python
def greet(name, *nicknames):
    print("Name:", name)
    print("Nicknames:", nicknames)

greet("Emil", "Em", "E-man", "maenu")
# other elements are taken as tuples and are printed all together
```

```python
def greet(name, **details):
    print("Name:", name)
    print("Details:", details)

greet("Emil", age=25, hobby="coding",sports = "badminton")
#  **this helps in making the variable as DICT
```

```python
# ARGS AND KWARGS TOGETHER 
def demo(a, *args, **kwargs):
    print("a:", a)
    print("args:", args)
    print("kwargs:", kwargs)

demo(1, 2, 3, x=4, y=5)
```

### Local and Global variable

```python
# local Variable

# Declared inside a function — usable only there

def demo():
    x = 10
    print(x)
```

```python
# Global Variable

# Declared outside — usable everywhere

x = 20

def demo():
    print(x)
```

```python
# how to change the global variable

x = 5

def change():
    global x
    x = 10

change()
print(x)
```

### Lambda Functions
for one line functions

```python
# Normal function:
def add(a, b):
    return a + b
add(3,2)
```

```python
# Lambda function
add = lambda a, b: a + b
print(add(2, 3))  # 5
```

```python
square = lambda x: x * x
print(square(4))  
```

```python
square = lambda x: x * x
print(square(5))
```

```python
nums = [1, 2, 3, 4]

squared = map(lambda x: x * x, nums)
print(list(squared))
```

```python
# filter() 
nums = [1, 2, 3, 4, 5, 6]

evens = filter(lambda x: x % 2 == 0, nums)
print(list(evens))
```

```python
students = [("Emil", 25), ("Anna", 20), ("John", 30)]

# sort by age
students.sort(key=lambda s: s[1])
print(students)
```
