### dict are unordered, mutable and dont allow duplicate values



```python
d1 = {
    "name" : "palak",
    "age" : 20,
    "score" : 90,
    "is adult" : True,                   #key can be string
     12.9 : 13,                          #float
     True : "correct",                   #boolean
     13 : False,                         #int
     "subjects" : ['english', 'arabic'], #list
     "values" :('dict',' list', 'tuple') #tuple
}
print (d1)
print(d1["subjects"])
print(d1["values"])
```

### nested dictionary

```python
student= { "name" : "rahul kumar" ,
           "subjects":
            { "phy": 98,
             "chem": 87,
             "bio":78}}
print(student)
print(student["subjects"])
print(student["subjects"]["chem"])
```

### Dict methods

```python
student= { "name" : "rahul kumar" ,
           "subjects":
            { "phy": 98,
             "chem": 87.9,
             "bio":78}}
print(student.keys())      #provide all the keys of outer layer
```

```python
student= { "name" : "palak" ,
           "subjects":
            { "phy": 98,
             "chem": 87.9,
             "bio":78}}
print(student.values())       #all the values would appear
print(list(student.values())) #to convert it into lists
```

```python
student= { "name" : "ARYAN" ,
           "subjects":
            { "phy": "PASS",
             "chem": "PASS",
             "bio":"FAIL"}}
print(student.items())         #pass the values as tuples
print(list(student.items()))   #converts to the list
pairs = list(student.items())
print(pairs[0])                 #to access first pair
```

```python
student= { "name" : "palak" ,
           "subjects":
            { "phy": 98,
             "chem": 87.9,
             "bio":78}}

print(student.get("name"))   #returns the value
print(student.get("name2"))  #if this key is not present it won't give ERROR
```

```python
student= { "name" : "palak" ,
           "subjects":
            { "phy": 98,
             "chem": 87.9,
             "bio":78}}
print(student.update({"city":"Delhi"}))  #update the dictionary with new values
print(student)
```

### merge two dict

```python
d1 = {1:'a',2:'b'}
d2 = {3:'c',4:'d'}
merged = {**d1, **d2}   #merging two dict
print(merged)
```

### find students who scored more than 80

```python
score = {
    "junaid" : 90,
    "ahmad" : 79,
    "rahul" : 93,
    "sophia" : 37,
    "riya" : 78,
    "sonal" : 93}
for name, marks in score.items():
    if marks > 80:
        print(name, ":", marks)
```
